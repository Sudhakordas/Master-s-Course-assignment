```python
df = pd.read_csv("Allfile.csv") # hear 'ALLfile.csv contain the mean'

# Making a list of tuple of time and uid for getting the mean value in a specific date and aid
uid_time = []
for uid,time in zip(df.uid, df.time):
    uid_time.append((uid,time))

#converting x,y,z axis valuer into list
x_axis_value = df.x_axis.to_list()
y_axis_value = df.y_axis.to_list()
z_axis_value = df.z_axis.to_list()
```

#  Function for finding mean value for each aid in a specific date

```python
mean_list = [] # this variable will hold  all the data

from operator import itemgetter
for key,value in user_task_info.items():
    final_result = {} # this dictionary wiil contrrain mean of all the aid in a specific date
    print("Start calculating mean for :" ,key)
    value = sorted(value,key=itemgetter(1))
    value = value + [(80,80)] # adding a extraa tuple to each list of touple for workine till the last touple
    print('All the (aid, uid) pair on ',key, 'are :', value)
    print('\n')
    
    x, y, z =  0,0,0   # intializing the x, y, z value
    count = 0 #this variabvle is for how many tims an aid is appear
    ad = value[0][1] # initializing the aid value
    aid_lis = [] # this variable will contain all the unique aid in a date
    all_mean_aid = {} # mean of all the unique aid
    
    for uid, aid in value:
        #This if condition is for storing unique aid
        if aid not in aid_lis:
            aid_lis.append(aid)
            
        #If respective two aid is same then try to get the x_mean, y_mean,z_mean from the list(converted above) on the specific datye and aid  
        if ad == aid:
            try:
                index = uid_time.index((uid,key))
                x_mean = (x_axis_value[index])
                y_mean = (y_axis_value[index])
                z_mean = (z_axis_value[index])
                #print('x,y,z', x_mean, y_mean, z_mean)
           #other wise set those value as 0
            except:
                x_mean = 0
                y_mean = 0
                z_mean = 0
           
        # then sum up the x,y,z value for two same aid
            x+= x_mean
            y+= y_mean
            z+= z_mean
            count+=1
            
        #If two aid is not same     
        else:
            mean_aid = {}
            mean = {}
            
            #calxulate the mean of previous aid 
            x_mean = round((x/count), 2)
            y_mean = round((y/count), 2)
            z_mean = round((z/count), 2)
            
            mean['mean'] = {
                "x_axis":x_mean,
                "y_axis":y_mean,
                "z_axis":z_mean  
                
            }
            ai = str(aid_lis[-2])
            mean_aid[ai] = mean
            all_mean_aid.update(mean_aid)
            
            print('Mean Dictionary for', aid_lis[-2],'and it is present ', count, 'times) ', 'is', mean_aid)
            print('Mean of', (aid_lis[-2]), 'on date', key, x_mean, y_mean, z_mean)
            print('\n')
            
            #Then initializing the x, y,z value for the now aid
            try:
                index = uid_time.index((uid,key))
                x = (x_axis_value[index])
                y = (y_axis_value[index])
                z = (z_axis_value[index])
            except:
                x = 0
                y = 0
                z = 0
                
            #set the new aid and the count as 1
            ad = aid
            count = 1  
            
    #this postion is for storing the mean value in a dictionary according to the submission file   
    
    final_result['date'] = key  #Storing date in a date variable
#     print('printing from dictionary :', final_result['date'])
#     print('printing the key :', key)
    print('printing all mean_id :', all_mean_aid)
    final_result.update(all_mean_aid) #updating the finale result for each date
    print('\n')
    print('final_result:', final_result)
    mean_list.append(final_result) # and finaly appending into the mainlist
    print('Done with with calculating mean for : ', key)
    print('\n')
```
